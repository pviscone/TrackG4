## Posizione degli hit del tipo di particella settata

```python
particle="mu-"
part_df=detector.Filter(f"ParticleID=={partDict[particle]}")
```

```python
for i in range(6):
    exec(f"hist2D{i}=part_df.Filter(f\"Layer=={i+1}\").Histo2D((\"hist\",\"hist\",numPix,-0.5,0.5,numPix,-0.5,0.5),\"XPixel\",\"YPixel\")")
    exec(f"hist2D{i}.SetTitle(\"XPixel vs YPixel: Layer{i+1}, Particle: {particle}\")")
    exec(f"hist2D{i}.SetXTitle(\"Position [m]\")")
    exec(f"hist2D{i}.SetYTitle(\"Position [m]\")")
    exec(f"hist2D{i}.SetZTitle(\"Counts\")")
    exec(f"c{i+2}=ROOT.TCanvas()")
    exec(f"hist2D{i}.Draw(\"colz\")")
    exec(f"c{i+2}.Draw()")

```

```python
c=ROOT.TCanvas()
hist=detector.Histo1D(("hist","hist",10,0,5),"ParticleID")
hist.Draw()
c.Draw()
```

## Energia depositata per step nel detector da tutte le particelle in GeV

```python
c=ROOT.TCanvas()
hist=detector.Histo1D(("hist","hist",40,0,1e-3),"EnergyDeposited")
hist.Draw()
c.Draw()
```
